### Apply

````python

def kategoryzuj_populacje(ludnosc):
  if ludnosc>10**6:
    return 'ogromne miasto'
  elif ludnosc>2 *10**5:
    return 'srednie miasto'
  else:
    return 'male miasto'
dane['Wielkosc']=dane.population.apply(kategoryzuj_populacje) #operacja na całej kolumnie
````

````python
# operacje na wierszu
def polozenie(wiersz):
  wiersz['polozenie']=f'szerokość: {wiersz.lat}, długość: {wiersz.lng}'
  return wiersz
dane.apply(polozenie,axis=1);
````

* wyodrębnij dane z roku 2010 i zadbaj o brakujące dane
* zaprezentuj na wykresie ludność w poszczególnych stanach (bez sumarycznej wartości w całym USA) w roku 2010 w podziale na kategorie wiekowe
* zaprezentuj na wykresie gęstości zaludnienia w roku 2010 w poszczególnych stanach (j.w.) w kolejności malejącej

````python
#Zadanie 2
zb1 = pd.read_csv('https://raw.githubusercontent.com/jakevdp/data-USstates/master/state-abbrevs.csv')
zb2 = pd.read_csv('https://raw.githubusercontent.com/jakevdp/data-USstates/master/state-areas.csv')
zb3 = pd.read_csv('https://raw.githubusercontent.com/jakevdp/data-USstates/master/state-population.csv')

zb2.drop_duplicates(inplace=True)
zb3.drop_duplicates(inplace=True)

zb12 = pd.merge(zb1, zb2, on='state', how='left')
zb123 = pd.merge(zb12, zb3, right_on='state/region', left_on='abbreviation')
zb123 #Dane bez USA

#dane z USA
zb21 = pd.concat([zb1, zb2])
zbUSA = pd.merge(zb12, zb3, right_on='state/region', left_on='abbreviation', how='right')
zbUSA
````

````python
#wyodrębnij dane z roku 2010 i zadbaj o brakujące dane
dane2010= zb123[zb123['year']==2010]
dane2010.set_index('state',inplace=True)
dane2010 # bez USA

dane2010USA = zbUSA[zbUSA['year']==2010]
dane2010USA
````

````python
#zaprezentuj na wykresie ludność w poszczególnych stanach (bez sumarycznej wartości w całym USA) w roku 2010 w podziale na kategorie wiekowe
dane_grupy=dane2010.groupby(by='ages')
dane_grupy['population'].plot()
plt.xticks(rotation=90)
````

````python
#zaprezentuj na wykresie gęstości zaludnienia w roku 2010 w poszczególnych stanach (j.w.) w kolejności malejącej
dane2010['gęstość']=dane2010['population']/dane2010['area (sq. mi)']
dane2010t = dane2010[dane2010['ages']=="total"]
dane2010t['gęstość'].sort_values(ascending=False).plot()
plt.xticks(rotation=90)
````
